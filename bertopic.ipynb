```python
import pandas as pd
import pandas as pd

df = pd.read_csv("/content/finfin2024.csv", encoding='ISO-8859-1')

docs = df['text'].tolist()
```

```python
from bertopic import BERTopic
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_info = topic_model.get_topic_info()
print(topic_info)

print(topic_model.get_topic(0))
doc_info = topic_model.get_document_info(docs)
print(doc_info)
```

```python

from bertopic.representation import KeyBERTInspired

representation_model = KeyBERTInspired()
topic_model = BERTopic(representation_model=representation_model)
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_info = topic_model.get_topic_info()
print(topic_info)
```

```python
topic_words = topic_model.get_topic(0)
print("Topic 0 words:", topic_words)
```

```python
for topic_id in range(3):
    print(f"Topic {topic_id}: {topic_model.get_topic(topic_id)}")
```

```python

topic_model.visualize_topics()
topic_model.visualize_barchart()
```

```python
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary
```

```python
topics_words = topic_model.get_topics()


texts = [doc.split() for doc in docs]
dictionary = Dictionary(texts)

# Converting the topics from BERTopic to a format suitable for gensim CoherenceModel
topic_words = []
for topic_id, words in topics_words.items():
    topic_words.append([word for word, _ in words])


coherence_model = CoherenceModel(topics=topic_words, texts=texts, dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()

print(f"Coherence Score: {coherence_score}")
```
